# **Polynomial?**

For example: y= b0 + b1x1 + b2x1^2 + .....+ bnx1^n
It is a modified case of multiple linear regression





## **Importing Libraries**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## **Importing The Dataset**

```python
dataset=pd.read_csv('Position_Salaries.csv')
# we will not take the first column as it is uniquely labeled on second column
x=dataset.iloc[:,1:-1].values
y=dataset.iloc[:,-1].values
```

## **Training the Linear Regression model on the whole dataset**

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(x,y)
```

## **Training the Plynomial Linear Regression Model on the whole dataset**

Our target is to make an matrix from 0 to n power of x. and then we will apply it on linear regression

PolynomialFeatures(degree=2) is a function from the sklearn.preprocessing module in Python's scikit-learn library. It is used to perform polynomial feature expansion, which is a technique in machine learning to add additional features derived from the original input data by raising them to a specified power or degree.

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
x_poly=poly_reg.fit_transform(x)
lin_reg_2=LinearRegression()
lin_reg_2.fit(x_poly,y)
```

## **Visualizing the Linear Regression model**

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title('truth or Bluff(Linear Regression Model)')
plt.xlabel('position label')
plt.ylabel('salary')
plt.show()
```

Not well adapted


```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(x_poly),color='blue')
plt.title('truth or Bluff(Polynomial Regression Model)')
plt.xlabel('position label')
plt.ylabel('salary')
plt.show()
```

Much better than before

## **Predicting a new result with Linear Regression**

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
